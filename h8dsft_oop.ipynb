Task : 

Create a library catalog system using OOP. The system should allow users to add books to the catalog, search for books by title or author, and remove books from the catalog.

Requirements:

1. Each book should have a title, author, and unique identifier

2. The system should provide the following functionality:

●Adding a book to the catalog.

●Searching for books by title or author.

●Removing a book from the catalog.

●Displaying all books in the catalog.

3. Do the exception handling.


```python
class LibraryBook:
    '''this statement is used to generate a new OOP which the LibraryBook written above is the class of the system'''
    
    def __init__(self, book, author, uniqueid): 
        '''this statement is used to generate a new function in which will generate an object attribute consists of the book (name), its author, and unique id'''
        self.book = book
        self.author = author
        self.uniqueid = uniqueid

class LibrarySystem: 
    '''this statement is used to generate a new class OOP which is the library system and it contains the whole library system asked in the task (adding books, removing books, displaying all the books, and searching for books)'''
    def __init__(self):
        self.books = []
        '''this statement used to generate new function that will contain a collection of statements into the self.books list'''

    def addbook(self, book):
        self.books.append(book)
        '''generating a new function to add the book'''

    def removebook(self, book):
        self.books.remove(book)
        '''generating a new function that will execute a command of removing a book'''

    def searchauthors(self, author):
        written_by_author = []
        for book in self.books:
            if book.author == author:
                written_by_author.append(book)
        return written_by_author
    '''generating new function to executing a command of searching a book based on its author'''

    def searchname(self, book_title):
        for book in self.books:
            if book.book == book_title:
                return book
    '''generating new function to executing a command of searching a book based on book title'''

    def searchbookuniqueid(self, uniqueid):
        for book in self.books:
            if book.uniqueid == uniqueid:
                return book
    '''generating new function to executing a command of searching a book unique id'''

    def librarycatalog(self):
        return self.books
    '''this statement is used to generate a command of printing the books list (books)'''

def main():

    lib = LibrarySystem()  
    '''this statement generates an instance of the LibrarySystem class'''

    print("=====================")
    print("Welcome to freeliby")
    print("=====================")
    
    '''these statements used to generate the header of the library system'''

    #these statements below used to generate a main menu display that will be appeared in the front page as options for users to add new books, search a book, remove a book, displaying the library catalog, and exiting from the library system
    while True:
        print("Choose an option below")
        print("1. Adding new books")
        print("2. Search a book by book title")
        print("3. Search a book by author")
        print("4. Removing a book")
        print("5. Library catalog")
        print("6. Exit")
    
        #this is a statement that will generate a function to allocate the user's input for accessing their desired system in this library
        try :
            option = int(input("Enter option here: "))
        except ValueError:
            print("your input is invalid. please type an integer number as your input")
            continue
    
    
        if option == 1:
            book_title = input("Enter book title: ")
            author = input("Enter author: ")
            uniqueid = input("Enter unique ID: ")
            book = LibraryBook(book_title, author, uniqueid)
            lib.addbook(book)
            print("Book added successfully!")

        elif option == 2:
            book_title = input("Type the book title: ")
            found_book = lib.searchname(book_title)
            if found_book:
                print(f"Book found: {found_book.book}, Author: {found_book.author}")
            else:
                print("Book not found")

        elif option == 3:
            book_author = input("Type the author: ")
            books_by_author = lib.searchauthors(book_author)
            if books_by_author:
                print(f"Books by {book_author}:")
                for book in books_by_author:
                    print(f"Book: {book.book}, Author: {book.author}")
            else:
                print("Books not found")

        elif option == 4:
            book_title = input("Enter book title to remove: ")
            found_book = lib.searchname(book_title)
            if found_book:
                try:
                    lib.removebook(found_book)
                    print(f"Book '{book_title}' removed successfully!")
                except ValueError :
                    print("an error occurred during the process of removing a book")
            else:
                print("Book not found!")

        elif option == 5:
            catalog = lib.librarycatalog()
            if catalog:
                print("Library Catalog:")
                for book in catalog:
                    print(f"Book: {book.book}, Author: {book.author}, Unique ID: {book.uniqueid}")
            else:
                print("The library catalog is empty!")

        elif option == 6:
            print("Have a good day")
            break

        else:
            print("Enter a valid input!")
        '''the ifs, elifs, and else used above are the conditions made to generate the desired system of the users, consists of 6 different input based on 6 different desired system, started with adding a new books, and ending with exiting the system'''
        
if __name__ == "__main__":
    main()
#these statement is used in case a unittest run is necessary
```
